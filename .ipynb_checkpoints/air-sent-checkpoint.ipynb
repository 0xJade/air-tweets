### I. ___The Whats___
---
#### _i. What is ULMFiT?_
- Universal Language Model Fine-tuning for Text Classification
- Uses transfer learning method
---
#### _ii. What are its benefits?_

  - Address NLP problems with small datasets/amount of computational resources
  - Make NLP problems easier
  - Out performs state of the art methods on 6 separate text classification tasks
  - Reduces error 18-24% on majority of datasets
  - With only 100 labeled samples, it matches performance of training from scratch - on 100x more data
---
#### _iii. What is Transfer Learning?_
- An ML problem that stores knowledge gained from solving one problem, and directly applying or transferring that knowledge to another related problem
- Common approach is by fine tuning original model
- Simple models use a single layer of weights (embeddings)

## II. ___The Whys___
---

#### a. ULMFiT - Why is it significant? & Findings

- Controlling how fast the model learns allows for better adaptability
- Learns well from a limited number of examples
- Performs better than traditional NLP methods
- Requires magnitude of 100X less data than current state of the art methods

## III. ___The Hows___
---------
#### a. Pretrain Language Model on general corpus
- Preprocess data, normalize, drop nas, use Average SWD Long Short Term Memory Model

#### b. Fine-tune new dataset
- Apply
    - Discriminative fine-tuning
        - Tunes each layer in neural network with different learning rates
        - Stochastic gradient descent accounts for another variable - theta with respect to the layer
    - Slanted triangular learning rate
        - For adapting params, we want the model to converge quickly
        - First linearly increases learning rate
        - Then linearly decays

````
# -----------------------------------
# import dependencies
# -----------------------------------
import io
import os
import re
import sys
import fastai
from fastai import *
from fastai.text import *
from functools import partial
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
import seaborn as sns
````

## I. Train Language Model
---

````
# -----------------------------------
# a. Load Data
# -----------------------------------
df = pd.read_csv('../Tweets.csv',encoding='ISO-8859-1')
# -----------------------------------
# b. Visualize Sentiment
# -----------------------------------
sns.set(style="darkgrid")
sns.set(rc={'figure.figsize':(15,7)})
sns.countplot(x="airline", hue='airline_sentiment', data=df)
plt.title("Airline Sentiment For Each Airline");
df = pd.DataFrame({'text': df.text, 'label': df.airline_sentiment})
````

````
# -----------------------------------
# c. Encode Categories as Int Type
# -----------------------------------
label_dict = {'negative': 1, 'neutral': 2, 'positive': 3}
df.label = df.label \
  .apply(lambda x: label_dict[x])
# -----------------------------------
# d. Preprocess - Clean
#    ==================
#    - Replace airline tag to reduce
#      error in model
#    - Check NA counts
# -----------------------------------
regex = r"@(VirginAmerica|united|SouthwestAir|Delta|USAirways|AmericanAir)"
df['text'] = df['text'].apply( \
  lambda x: re.sub(regex, '@airline', x, flags=re.IGNORECASE))
df.text = df.text.str.replace("[^a-zA-Z]", " ")
````

````
# -----------------------------------
# d. Check Distribution of Labels to
#    Reduce Bias
# -----------------------------------
print('Airline Sentiment Value Counts\n' + 
      f'{df.label.value_counts()}')
print('\nText NaN Counts\n' + 
      f'{df.isna().sum()}')
````

````
# -----------------------------------
# d. Create a More Even Distribution
#    of Label Types for Model
#    ===============================
# 
#      - 3X as more negative labels 
#        compared to neutral and
#        positive
# 
#      - Randomly sample 3,000 negative
#        labels to train model with more
#        fair representation of data!!!
# 
# -----------------------------------
df = pd.concat([ df[ df.label.isin( [2, 3] ) ], \
                 df[ df.label==1].sample(3000) ])
df = df.reset_index(drop=True)
df.label.value_counts()
````

````
# -----------------------------------
# e. Tokenize
# -----------------------------------
tokenized = df.text.apply(lambda x: x.split())
# -----------------------------------
# f. Remove Stopwords
# -----------------------------------
stop_words = stopwords.words('english')
tokenized = tokenized.apply(lambda x: \
  [item for item in x if item.lower() not in stop_words] )
# -----------------------------------
# f. De-tokenize
# -----------------------------------
detokenized = []
for i in range(len(df)):
  toke = ' '.join(tokenized[i])
  detokenized.append(toke)
df.text = detokenized
````

````
# -----------------------------------
# g. Break into Train and Validation
#    Datasets - 60:40
# -----------------------------------
train, val = train_test_split(
    df, \
    stratify = df.label,
    test_size=0.4,
    random_state=10 )
# -----------------------------------
# h. Transform dataset in DataBunch
#    for classification to train an
#    RNN classifier
# -----------------------------------
lm_data = TextLMDataBunch.from_df(train_df = train, \
                                  valid_df = val, path='')
clas_data = TextClasDataBunch.from_df(
    train_df = train, \
    valid_df =  val, \
    vocab = lm_data.train_ds.vocab, \
    bs = 32, path = '')
````

# _Chapter 2 - ASGD Weight Dropped LSTM_
---
[ASGD Weight Dropped LSTM](https://arxiv.org/abs/1708.02182)

## I. The Whats

- What Language Model Should we Use?

  - ULMFiT suggests using ASGD Weight Dropped LSTM for our pretrained model in step 1

- What is ASGD Weight-Dropped LSTM?

  - It's a popular Long Short Term Memory Language Model (type of Recurrent Neural Network) that has a unique set of regularization techniques to improve performance

  - Uses DropConnect on hidden-to-hidden weights as form of recurrent regularization

- What is a Recurrent Neural Network?

  - Networks with loops to allow for persisting information by passing it from one step of the network to the next

  - If you think of the human brain, we don't make decisions by starting from scratch. Rather, we use our past experiences and knowledge to pass judgement and carry out our decision making process

  - Can be thought of multiple copies of the same network, passing message to each successor

  - RNNs are capable of long term dependencies

  - However, there's a point where this gap between relevant information and point where more context is needed grows to be too large for RNNs
    - To a point where they are unable to learn and make connections

  - This is where Long Shert Term Memory networks come into play!

- What is LSTM?

  - Special kind of RNN capable of learning long term dependencies

  - Introduced by Hochreiter & Schmidhuber (1997)
    - http://www.bioinf.jku.at/publications/older/2604.pdf

  - What sets LSTM apart from a regular RNN?
    - RNNs have form of a chain of repeating modules - with a single tanh layer for the structure
    - Whereas the LSTM has 4 neural network layers of interaction.

  - Key to LSTM is cell state - horizontal line running through the entire chain of layers

  - Has ability to add/remove information to cell state with structures called gates

  - Gates regulate information traffic



  - Composed of a sigmoid neural net layer and a pointwise multiplication operator

  - Sigmoid layer outputs number 0-1 describing information permeability

  - 0 means:

  ![](https://media.giphy.com/media/njYrp176NQsHS/giphy.gif)


  - 1 means:

  ![](https://media.giphy.com/media/lHQuX9d5DBhug/giphy.gif)


- What is ASGD Weight Dropped LSTM?

  - RNN recurrent connections are prone to overfitting

  - Dropouts have been very successful in feed forward and convulational neural networks

  - However, applying dropouts similarly to RNN's hidden state is ineffective, disrupts long term retention ability

  - To counter this problem, DropConnect is proposed

  - Typically in Dropout, randomly selected subset of activations is set to zero within each layer

  - DropConnect, instead of activations, randomly selected subset of weights within the network is set to zero

  - Each unit receives input from random subset of units in previous layer

## II. The Whys

- Why should we use ASGD Weight Dropped LSTM?

  - Because it's memory retention and problem solving skills are modeled directly after the human brain

  - Regulating relevant information to the context of a sentence allows a model to make more informed decisions

  - It has 19 syllables in the title and sounds fancy as heck! (just kidding ;D)


## III. The How

  - How does the Average SGD algorithm work?

    - It takes gradient update step similarly to SGD

    - Instead of returning weight computed in current iteration it takes into account weights of previous iterations

    - Returns an average

    ```
    # -----------------------------------
    # SGD Weight Update
    # -----------------------------------
    w_t = w_prev - lr_t * grad(w_prev)

    # -----------------------------------
    # Average SGD Weight Update
    # -----------------------------------
    avg_fact = 1 / max(t - K, 1)

    if avg_fact != 1:
      w_t = avg_fact * (sum(w_prevs) + (w_prev - lr_t * grad(w_prev)))
    else:
      w_t = w_prev - lr_t * grad(w_prev)
    ```
    - Where
      - K = min num iterations run before weight average starts
      - t = current number of itertions
      - sum(w_prevs) = sum of weights from iteration K to t
      - lr_t learning rate at iteration t

  - Extended Regularization Techniques

    - Variable Length Backpropagation Sequences

      - Fixed length backprogation through (BPTT) is inefficient

      - Use of variable length backpropagation sequences is proposed

      - First select length to be bptt with probability p and bptt/2 with proabaility 1-p

      - Pytorch implementation:

      ```
      base_bptt = bptt if np.random.random() < 0.95 else bptt / 2
      ```

      - base_bptt is used to retrieve seq_len
      - s = standard deviation
      - N = normal distribution

      ```
      seq_len = max(5, int(np.random.normal(base_bptt, 5)))
      ```
      - learning rate is rescaled based on the seq_length

      ```
      lr2 = lr * seq_len / bptt
      ```

    - Variational Dropout

      - In standard dropout, new dropout mask sampled each time drouput connection is called
      - Variational has dropout mask sampled only upon first call
        - locked dropout mask is repeatedly used for all connections within forward and backward pass

      - while DropConnect is used to regularize hidden to hidden transition within RNN
      - Variational Dropout is used for all other dropout operations

    - Activation Regularization

      - L2 regularization is often used on weights to reduce overfitting
      - L2 decay can also be used on activations

# Chapter 3 - Fine Tuning!

[Hyper Parameter Fine Tuning - Learning Rate, Momentum, Weight Decay](https://arxiv.org/pdf/1803.09820.pdf)

````
# -----------------------------------
# a. Getting Hyped about Hyper Params
#    ================================
# 
#      - One cycle and cyclic momentum
#        allows model to be trained on
#        higher learning rate - converge
#        more quickly
# 
#      - However, we want to avoid
#        learning too quickly, so
#        we'll vary our LRs by using
#        learn.find_lr & experimenting
# 
#      - The art of adjusting hyper params
#        can be tricky. But with benefits
#        such as reducing training time
#        and performance improvement it's
#        something we don't want to skip
# 
#      - Balance of underfitting vs overfitting
# 
#         - Requires examining training's
#           test/validation loss for
#           underfitting/overfitting
# 
# -----------------------------------
lrs = {
    0: 0.001,
    1: 0.01,
    2: 0.1
}; wds = {
    0: 1e-5,
    1: 1e-4,
    2: 3e-4
}; moms = {
    0: (0.8, 0.7),
    1: (0.8, 0.6),
    2: (0.8,0.5)
}
# -----------------------------------
#   i. Test LR=0.1
# -----------------------------------
learn.fit_one_cycle(1, max_lr=0.1, moms=moms[0], \
                        wd=wds[0], tot_epochs=3)
# for i in range( 0, len(lrs) ):
#   for j in range( 0, len(lrs) ):
#     print('-----------------------------------')
#     print(f'lr=0.1, moms={moms[i]}, wd={wds[j]}')
#     print('-----------------------------------')
    
#     print('\n')
#     learn.recorder.plot_losses()
#     plt.grid()
#     plt.title(f'Loss vs. Epoch\nlr=0.1, moms={moms[i]}, wd={wds[j]}')
````

````
# -----------------------------------
#   ii. test LR=0.01
# -----------------------------------
for i in range( 0, len(lrs) ):
  for j in range( 0, len(lrs) ):
    print('-----------------------------------')
    print(f'lr=0.01, moms={moms[i]}, wd={wds[i]}')
    print('-----------------------------------')
    learn.fit_one_cycle(1, max_lr=0.1, moms=moms[i], \
                        wd=wds[j], tot_epochs=100 )
    learn.recorder.plot_losses()
    plt.grid()
    plt.figure(figsize=(9,3.0))
    plt.title(f'Loss vs. Epoch\nlr=0.01, moms={moms[i]}, wd={wds[j]}')
````

````
# -----------------------------------
#   iii. test LR=0.001
# -----------------------------------
for i in range( 0, len(lrs) ):
  for j in range( 0, len(lrs) ):  
    print('-----------------------------------')
    print(f'lr=0.01, moms={moms[i]}, wd={wds[i]}')
    print('-----------------------------------')
    learn.fit_one_cycle(1, max_lr=0.1, moms=moms[i], \
                        wd=wds[j], tot_epochs=100 )
    learn.recorder.plot_losses()
    plt.grid()
    plt.title(f'Loss vs. Epoch\nlr=0.001, moms={moms[i]}, wd={wds[j]}')
    print('\n\n')
````

## III. Classify & Predict!

````
# -----------------------------------
# a. Save Encoder, Learn w/ Classifier
#    ================================
# 
#      - One cycle and cyclic momentum
#        allows model to be trained on
#        higher learning rate - converge
#        more quickly
# 
#      - However, we want to avoid
#        learning too quickly, so
#        we'll vary our LRs by using
#        learn.find_lr & experimenting
# 
#      - The art of adjusting hyper params
#        can be tricky. But with benefits
#        such as reducing training time
#        and performance improvement it's
#        something we don't want to skip
# 
# -----------------------------------
learn.save_encoder('ft_enc')
learn = text_classifier_learner(clas_data, arch=AWD_LSTM, drop_mult=0.5)
learn.load_encoder('ft_enc')
learn.freeze()
learn.lr_find()
learn.recorder.plot()
````

````
learn.freeze_to(-2)
lr /= 2
learn.fit_one_cycle(1, slice(lr/(2.6**4), lr), moms=moms, wd=wd)
````

````
learn.freeze_to(-3)
lr /= 2
learn.fit_one_cycle(1, slice(lr/(2.6**4), lr), moms=moms, wd=wd)
````

````
learn.unfreeze()
lr /= 5
learn.fit_one_cycle(2, slice(lr/(2.6**4), lr), moms=moms, wd=wd)
````
